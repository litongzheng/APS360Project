```python
sp_text = ""

chars = []
i = 0
for char in open('character.txt'):
    chars.append(char.strip() + ":\t")
    if i>=300:
        break
    i = i+1
lines = [] 

i = 0
for line in open('line.txt'):
        lines.append(line)
        if i>=300:
            break
        i = i+1
        
for i in range(len(lines)):
    total = chars[i] + lines[i]
    sp_text += total
    
print(sp_text)
```
